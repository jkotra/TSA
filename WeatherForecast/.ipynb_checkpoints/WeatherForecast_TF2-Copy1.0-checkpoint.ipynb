```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')

from sklearn.preprocessing import MinMaxScaler as minmax
from sklearn.preprocessing import LabelEncoder

import tensorflow as tf
```

```python
tf.__version__
```

```python
data = pd.read_csv("data/nyc.csv",parse_dates=['Date'],index_col='Date')
data.head()
```

```python
train_cols = ["Mean.TemperatureF","Mean.Humidity","season"]
data = data[train_cols]
#Label encode season.
le = LabelEncoder()
data['season'] = le.fit_transform(data['season'])
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
#Visualize our data

plt.figure(dpi=200,figsize=(17,7))
data["Mean.TemperatureF"].plot(legend=True,color='orange')
plt.title('Temperature')
plt.ylabel("Fahrenheit - F")
plt.show()
```

```python
#Visualize our data - 2

plt.figure(dpi=200,figsize=(17,7))
data["Mean.TemperatureF"][:'2014'].plot(legend=True,color='orange')
data["Mean.TemperatureF"]['2014':].plot(legend=True,color='green')
plt.legend(['Training set (Before 2014)','Test set (2014 and beyond)'])
plt.title('Mean Temp.')
plt.ylabel("F")
plt.show()
```

```python
def make_dataset_multivariant_many_to_one(array,time_steps,target):

    x = []
    y = []

    for i in range(time_steps,len(array)):
        x.append(array[i-time_steps:i])
        y.append(array[i][target])
        
    return np.array(x), np.array(y)
```

```python
train = data[:"2014"].values

#scale values between 0 and 1
x_scaler = minmax(feature_range=(0,1))
y_scaler = minmax(feature_range=(0,1))


y_scaler.fit_transform(train[:,0].reshape(-1, 1))
train = x_scaler.fit_transform(train)

trn_x,trn_y = make_dataset_multivariant_many_to_one(train,60,1)
```

## The three dimensions of this input are:

* **Samples**. One sequence is one sample. A batch is comprised of one or more samples.
* **Time Steps**. One time step is one point of observation in the sample.
* **Features**. One feature is one observation at a time step.

```python
trn_x = trn_x.reshape(-1,60,3)

print(trn_x.shape,trn_y.shape)
```

### Model Architecture

* LSTM(100)
* LSTM(100)
* LSTM(100)
* LSTM(100)
* Dense(500)
* Dropout(0.2)
* Dense(1)

```python
model = tf.keras.models.Sequential()

model.add(tf.compat.v1.keras.layers.CuDNNLSTM(100, return_sequences=True))
model.add(tf.compat.v1.keras.layers.CuDNNLSTM(100, return_sequences=True))
model.add(tf.compat.v1.keras.layers.CuDNNLSTM(100, return_sequences=True))
model.add(tf.compat.v1.keras.layers.CuDNNLSTM(100))

model.add(tf.keras.layers.Dense(500))

model.add(tf.keras.layers.Dropout(0.2))

model.add(tf.keras.layers.Dense(1))

early_stop = tf.keras.callbacks.EarlyStopping(monitor='mse', patience=15)
model.compile(optimizer='adam', loss='mse', callbacks=[early_stop])
```

```python
model.fit(trn_x, trn_y, epochs=300)
```

## Validation

* Test our model on whole data including test set(2014 and beyond)

```python
total = x_scaler.transform(data.values)
total_x,total_y = make_dataset_multivariant_many_to_one(total,60,1)
```

```python
predicted = model.predict(total_x)
```

```python
predicted = y_scaler.inverse_transform(predicted)
```

```python
#make a dataset with real and predicted values.

infered = pd.DataFrame(data.iloc[60:])
infered['MT_Predicted'] = predicted
```

```python
infered.tail()
```

```python
plt.figure(dpi=200,figsize=(17,7))
infered["Mean.TemperatureF"].plot(legend=True,color='orange')
infered["MT_Predicted"].plot(legend=True,color='green')
plt.title('Temperature')
plt.ylabel("Fahrenheit - F")
plt.show()
```

## Testset visualization

```python
plt.figure(dpi=200,figsize=(17,7))
infered["2014":]["Mean.TemperatureF"].plot(legend=True,color='orange')
infered["2014":]["MT_Predicted"].plot(legend=True,color='red')
plt.title('Temperature')
plt.ylabel("Fahrenheit - F")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(infered["Mean.TemperatureF"],infered["MT_Predicted"])
print("RMSE:", np.sqrt(mse))
```
